```python
x = np.arange(-5, 5, 0.1)
y1 = x**2 - 2
y2 = 20 * np.cos(x)**2

plt.figure(1) # 1번 그래프 창 생성
plt.plot(x, y1)

plt.figure(2) # 2번 그래프 창 생성
plt.plot(x, y2)

plt.figure(1) # 1번 그래프 창 지정
plt.plot(x, y2)

plt.figure(2) # 2번 그래프 창 생성
plt.clf()  # 모든 그래프 지움
plt.plot(x, y1)
```

```python
# 하나의 그래프 창을 하위 그래프 영역으로 나누기 subplot(m,n,p) -> m x n 행렬에서 p는 왼쪽->오른쪽, 위-> 아래 차례로 행렬의 위치 지정
x = np.arange(0, 10, 0.1)
y1 = 0.3 * (x-5)**2 + 1
y2 = -1.5 * x + 3
y3 = np.sin(x)
y4 = 10*np.exp(-x) + 1

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)
```

```python
# 그래프 출력 범위 지정

x = np.linspace(-4, 4, 100)
y1 = x**3
y2 = 10 * x**2 - 2

plt.plot(x, y1, x, y2)
```

```python
# 그래프 꾸미기
import numpy as np
x = np.arange(0,5,1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3

plt.plot(x, y1, x, y2, x, y3, x, y4)
```

```python
plt.plot(x, y1, ">-r", x, y2, "s-g", x, y3, "d:b", x, y4, "-.Xc") # 컬러, 선 스타일, 마커 혼합 지정
```

```python
plt.plot(x, y1, "o", x, y2, "^", x, y3, "s", x, y4, "d") # 데이터의 각 점을 선으로 연결하는 것이 아니라 다양한 모양의 마커로 표시
```

```python
import numpy as np
from matplotlib import font_manager, rc

city = ['seoul', 'Incheon', 'Deajeon', 'DeaGu', 'Ulsan', 'Busan', 'GwangJu']

# 위도와 경도
lat = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]
lon = [126.97, 126.7, 127.38, 128.60, 129.31, 129.07, 126.85]

# 인구 밀도
pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

size = np.array(pop_den)
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y'] # 데이터별로 마커의 컬러 지정

plt.scatter(lon, lat, s=size, c=colors, alpha=0.4) # 인구밀도 별로 마커 크기 지정, alpha는 투명도
plt.xlabel("lon")
plt.ylabel("lat")
plt.title("pop_density")

for x,y, name in zip(lon, lat, city):
    plt.text(x,y,name) # 위도 경도에 맞게 도시 이름 출력

plt.show()
```

```python
## 막대그래프 bar

memberID = ["m_1", "m_2", "m_3", "m_4"] # 회원 ID
before_ex = [27, 35, 40, 33] # 운동 전
after_ex = [30, 38, 42, 37] # 후
```

```python
n_data = len(memberID)
index = np.arange(n_data) # 배열 생성(0,1,2,3)
plt.bar(index, before_ex) # bar(x,y)에서 x=index, y=before_ex로 지정
```

```python
plt.bar(index, before_ex, tick_label = memberID) # x축의 tick라벨을 원하는 데이터로 지정
```

```python
colors = ['r', 'g', 'b', 'm']
plt.bar(index, before_ex, color= colors, tick_label = memberID) # 컬러지정
```

```python
plt.bar(index, before_ex, tick_label = memberID, width=0.6) # 막대의 폭 변경
```

```python
plt.barh(index, before_ex, color= colors, tick_label = memberID) # 가로 막대 (barh)
```

```python
barWidth = 0.4
plt.bar(index, before_ex, align = "edge", color = 'c',  width=barWidth, label = "before" )
plt.bar(index + barWidth, after_ex, align = "edge", color = 'g', width=barWidth, label = "after" ) # After는 Before 옆으로 지정
plt.legend()
plt.xlabel("회원ID")
plt.ylabel("윗몸 횟수")
plt.title("윗몸 일으키기 운동 횟수 변화")
```
